```
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn import datasets
from sklearn.metrics import mean_squared_error
from sklearn import linear_model
```

**Train Dataset**

```
columnNames = ['age', 'workclass', 'fnlwgt', 'education', 'educational-num','marital-status', 'occupation', 'relationship', 'race', 'gender','capital-gain', 'capital-loss', 'hours-per-week', 'native-country','income']
urlTrain= "https://raw.githubusercontent.com/adityaaggarwal19/Data-Mining/master/adult_dataset.csv" 
trainData = pd.read_csv(urlTrain,header=None,names=columnNames)
trainData.reset_index(inplace = True, drop = True)
trainData.head()
```

**Test Dataset**

```
urlTest = "https://raw.githubusercontent.com/adityaaggarwal19/Data-Mining/master/adult_testset.csv" 
testData = pd.read_csv(urlTest,header=None,names=columnNames)
testData['income'].replace(regex=True,inplace=True,to_replace=r'\.',value=r'')
testData.reset_index(inplace = True, drop = True)
testData.head()
```

```
various_education_levels = {}
for x,y in trainData[['educational-num','education']].drop_duplicates().itertuples(index=False):
    various_education_levels[y] = x

education = round(pd.crosstab(trainData.education, trainData.income).div(pd.crosstab(trainData.education, trainData.income).apply(sum,1),0),2)
education = education.reindex(sorted(various_education_levels, key=various_education_levels.get, reverse=False))

bar_graph_education_income = education.plot(kind ='bar', title = 'Education Level vs Income Proportion', figsize = (10,8))
bar_graph_education_income.set_xlabel('Education level of instance')
bar_graph_education_income.set_ylabel('Income Proportion of dataset')
```

```
hours_per_week = round(pd.crosstab(trainData['hours-per-week'], trainData.income).div(pd.crosstab(trainData['hours-per-week'], trainData.income).apply(sum,1),0),2)
bar_graph_hours_per_week_income = hours_per_week.plot(kind ='bar', title = 'Proportion distribution across Hours per week', figsize = (20,12))
bar_graph_hours_per_week_income.set_xlabel('Hours per week')
bar_graph_hours_per_week_income.set_ylabel('Proportion of population')
```

**WORKING OF A SIMPLE CLASSIFICATION MODEL :-**


For understanding the working of classification algorithm let us take a simple set of rules to decide whether the salary of an individual is greater than 50k or less than or equal to 50k.

So there are two classes :-
  1. \>50k
  2. <=50k

Rules for a simple model can be :- 
  1. If { 
          (age >= 25) &&
          (education == masters || prof-school || doctrate) && 
          (marital_status == married) &&
          (hours_per_week >= 60)
        }<br>
        then : CLASS ASSIGNED BY MODEL IS >50k.
        <br>
        <br>

  2. If { 
          (age < 25) &&
          (workclass == never_worked || workclass == without_pay) &&
          (race == black || race == white || race == other)
        }<br>
        then : CLASS ASSIGNED BY MODEL IS <=50k.
        <br>
        <br>

Different models can classify with different accuracy, so while building a classification model we have to read the data as carefully as we can and apply more and more attributes as we can to classify a random record.
